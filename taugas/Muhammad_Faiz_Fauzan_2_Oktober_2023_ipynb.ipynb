```python
def kuadrat(x):
  return x**2
```

```python
kuadrat(-3)
```

```python
print(kuadrat(-2))
print(kuadrat(-1))
print(kuadrat(0))
print(kuadrat(1))
print(kuadrat(2))
print(kuadrat(3))

```

```python
import matplotlib.pyplot as plt
```

```python
x=[-2,-1,0,1,2,3]
y=[4,1,0,1,4,9]
plt.plot(x,y,'.')
```

```python

plt.plot(x,y)
'''INI GRAFIK GARIS DARI X DAN Y'''
```

```python
plt.plot(x,y,'g', label="Persamaan kuadrat")
plt.title("Gambar grafik x^2", color="red")
plt.xlabel("Nilai x")
plt.ylabel("Nilai y")
plt.legend(loc="lower right")
plt.show()
```
