```python
def fungsi(x):
  return 1/x
```

```python
def fungsi2(x):
  return (x**2-1)/(x-1)
```

```python
print(fungsi2(1.000001))
print(fungsi2(0.999999))
```

```python
def fungsi3(x):
  return abs(x)/x
```

```python
import matplotlib.pyplot as plt
import numpy as np

x = np.arange(-5,5,0.1)
y = fungsi3(x)
plt.plot(x,y)
plt.show()
```
